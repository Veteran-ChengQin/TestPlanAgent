```python
import os
import json

# 查看当前的工作目录
# current_path = os.getcwd()
# print(current_path)
# 定义根目录
root_dir = 'test_project/SWELancer-Benchmark/issues'
# root_dir = os.path.join(current_path, project_dir)

# 存储所有读取的 issue_data.json 文件内容
issue_data_list = []

# 遍历根目录下的所有目录
for dir_name in os.listdir(root_dir):
    dir_path = os.path.join(root_dir, dir_name)
    if os.path.isdir(dir_path):  # 检查是否为目录
        json_file_path = os.path.join(dir_path, 'issue_data.json')
        if os.path.isfile(json_file_path):  # 检查 issue_data.json 是否存在
            with open(json_file_path, 'r') as json_file:
                issue_data = json.load(json_file)
                issue_data_list.append(issue_data)

# 输出读取的 issue_data.json 文件内容
issue_data_list.sort(key=lambda x: x['price'], reverse=True)
issue_data_list
```

```python
with open('test_project/SWELancer-Benchmark/issue_data.json', 'w') as json_file:
    json.dump(issue_data_list, json_file)
```

```python
url_prefix = "https://github.com/Expensify/App/issues/"
issue_url_list = []
for info in issue_data_list:
    issue_id = info['_issue_id'].split('_')[0]
    issue_url = os.path.join(url_prefix, issue_id)
    issue_url_list.append(issue_url)

with open("data/issue_url_list.txt", 'w') as file:
    for url in issue_url_list:
        file.write(url + '\n')
```
